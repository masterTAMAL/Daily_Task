```python
#importing datashet
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
beast_cancer=sklearn.datasets.load_breast_cancer()#loading the datshet
```

```python
beast_cancer
```

```python
#loading data into dataframe
data_frame=pd.DataFrame(beast_cancer.data,columns=beast_cancer.feature_names)
```

```python
data_frame.head()
```

```python
#adding the target column
data_frame['label']=beast_cancer.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
#finding missing value
data_frame.isnull().sum()
```

```python
# statistical measure
data_frame.describe()
```

```python
#checking the distribution oftaret value
data_frame['label'].value_counts()
```

1--> Bengin
0-->Maligant


```python
data_frame.groupby('label').mean()
```

```python
#separating dataframe
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(y)
```

```python
#spllitng data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
model=LogisticRegression()
```

```python
#trainnng the logistic regression model using traing data
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
print('accuracy on training data:',training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_prediction)
print('accuracy on test data:',test_data_accuracy)
```
