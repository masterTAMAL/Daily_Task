```python
#Importing dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the csv data to a pandas dataframe
df=pd.read_csv('/content/heart_disease_data.csv')
```

```python
#checking the target
df['target'].value_counts()
```

```python
#spliting the dataframe
X=df.drop('target',axis=1)
Y=df['target']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
#logistic regression
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
# #model evalution
# model.score(X_train,Y_train)
model_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(model_prediction,Y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
model_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(model_prediction,Y_test)
```

```python
print('Accuracy on test data :',test_data_accuracy)
```

```python
#Building a Predictive value
input_data=(10000,1,0,200,154,2,18,182,15,1,40,25,3)
input_data_as_numpy=np.asarray(input_data)
input_data_reshape=input_data_as_numpy.reshape(1,-1)
predicted=model.predict(input_data_reshape)
print(predicted)
if(predicted[0]==0):
  print('The person does not have a heart disease')
else:
  print('The person has a heart disease')
```
